```python
#visual representation of dataset
import matplotlib.pyplot as plt

df[['NH4', 'BSK5', 'Suspended', 'O2', 'NO3', 'NO2', 'SO4', 'PO4', 'CL']].hist(
    figsize=(12, 10), bins=15, edgecolor='black'
)
plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Feature and target selection - Feature - independent variable and Target dependent variable
X = df[['id', 'year']]
y = df[pollutants]
```

```python
# Encoding - onehotencoder - 22 stations - 1 - 1
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
# Train, Test and Split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

```python
# Train the model
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(X_train, y_train)
```

```python
# Evaluate model
y_pred = model.predict(X_test)
```

```python
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('   MSE:', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print('   R2:', r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

```python
station_id = '22'
year_input = 2024

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

# Align with training feature columns
missing_cols = set(X_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded[X_encoded.columns]  # reorder columns

# Predict pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```

```python
def classify_water(pred):
    # Safe fallback if any expected key is missing
    def get(k): return pred.get(k, 0)

    O2 = get('O2')
    NO3 = get('NO3')
    NO2 = get('NO2')
    PO4 = get('PO4')
    CL  = get('CL')

    # Water Quality Category (fallback logic without NH4)
    if O2 >= 6 and NO3 < 10:
        quality = 'Good'
    elif O2 >= 4 and NO3 < 25:
        quality = 'Moderate'
    else:
        quality = 'Polluted'

    # Drinkability (fallback logic)
    if NO3 <= 50 and CL <= 250 and PO4 <= 0.3 and NO2 <= 0.1:
        drinkable = 'Safe'
    else:
        drinkable = 'Unsafe'

    return quality, drinkable
```

```python
pred_dict = dict(zip(pollutants, predicted_pollutants)) 
```

```python
quality_label, drinkability_status = classify_water(pred_dict)

# Print the results
print("Water Classification:")
print(f"  Water Quality : {quality_label}")
print(f"  Drinkability  : {drinkability_status}")
```
